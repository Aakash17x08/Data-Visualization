**plotly** is a Python library which is used to design graphs, especially interactive graphs. It can plot various graphs and charts like histogram, barplot, boxplot, spreadplot and many more. It is mainly used in data analysis as well as financial analysis. plotly is an interactive visualization library.

**cufflinks** connects plotly with pandas to create graphs and charts of dataframes directly.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

```

```python
#graph_object
#express
import plotly.graph_objects as go
import plotly.express as px
```

```python
#using graph object
#add_trace => adding data
#scatter plot
fig = go.Figure()
fig.add_trace(go.Scatter(x=[1,2,3,4,5], y=[1,7,8,2,6], mode = 'markers'))
fig.show()
```

```python
# making a line chart
fig = go.Figure()
fig.add_trace(go.Scatter(x=[1,2,3,4,5], y=[1,7,8,2,6], mode = 'lines'))
fig.show()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=tips['total_bill'],
        y=tips['tip'],
        mode='markers'
    )
)

fig.show()
```

```python
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=tips['total_bill'],
        y=tips['tip'],
        mode='lines'
    )
)

fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x = [1, 2, 3, 4, 5], y = [3, 8, 5, 2, 6]))
```

```python
fig = go.Figure()
fig.add_trace(go.Histogram(x = tips['total_bill']))
```

```python
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=tips['total_bill'],
        y=tips['tip'],
        mode='markers', marker_size = 5*tips['size']
    )
)

fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x = tips.total_bill, y = tips.tip, mode = 'markers', marker_size = 10*tips['size'], z = tips['size']))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x = [1, 2, 3, 4, 5], y = [3, 8, 5, 2, 6], z = [1, 2, 7, 3, 5]))
```

```python
#using plotly express px
x = [1, 2, 3, 4]
y = [10, 13, 15, 18]
fig = px.scatter(x, y)
fig.show()
```

```python
fig = px.line(x, y)
fig.show()
```

```python
x = ['A', 'B', 'C']
y = [10, 5, 6]
fig = px.bar(x = x, y = y)
fig.show()
```

```python
x = [5, 10, 3, 4, 5, 5, 6, 6, 10, 10, 3, 11, 11, 12, 3, 4,]
fig = px.histogram(x)
fig.show()
```

```python
x = ['A', 'B', 'C']
y = [5, 10, 3, 25, 12, 33, 100]
fig = px.box(x,y)
fig.show()
```

```python
#plotting heatmap
data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
fig = px.imshow(data)
fig.show()
```

```python
#bubble chart
x = [1, 2, 3, 4]
y = [5, 8, 12, 3]
fig = px.scatter(x, y, size = [30, 10, 50, 100])
fig.show()
```

```python
#3d chart
x = [1, 2, 3, 4]
y = [5, 8, 12, 3]
z = [3, 4, 5, 6]
fig = px.scatter_3d(x=x, y=y, z=z)
fig.show()
```

```python
#contour plot
x = [1, 2, 3, 4, 3, 23, 80]
y = [5, 8, 12, 3, 12, 35, 60]
fig = px.density_contour(x, y)
fig.show()
```
