```python
import seaborn as sns
import numpy as np
```

# **lmplot()**
regression plot is used in machine learning so we will not fully discuss what, how and where it is used but we will see how it is created and what are the values we can put inside it

```python
tips = sns.load_dataset('tips')
tips
```

## Example Usage:

```python
# Basic lmplot
sns.lmplot(data=tips, x='total_bill', y='tip')
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue = 'sex')
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue = 'sex', palette='coolwarm')
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue = 'sex', palette='coolwarm', markers=['o','v'])
```
