## ðŸ“Œ Plot â†’ Purpose (One-liners)

* **histplot** â†’ Distribution of numeric data
* **kdeplot** â†’ Smooth density curve
* **countplot** â†’ Count of categories
* **boxplot** â†’ Spread, median, outliers
* **violinplot** â†’ Distribution + density
* **scatterplot** â†’ Relationship between numbers
* **jointplot** â†’ Relationship + distributions
* **pairplot** â†’ All relationships at once

## ðŸ“Œ Golden Rule (Very Important)

> **Quantitative (numerical) data always goes on the Y-axis**
> **Qualitative (categorical) data goes on the X-axis**

---

## ðŸ”¹ Why this rule?

* X-axis â†’ categories / groups
* Y-axis â†’ values that can be measured and compared

---

### âœ… Example 1: Gender vs Marks
