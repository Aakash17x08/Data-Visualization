We use distribution plots to the quantitative data (categorical,non-numeric data)
### üìä Distribution Plots

Used to visualize the **distribution of quantitative data**.

---

### üìà Types of Distribution Plots

* **distplot / histplot**
* **jointplot**
* **pairplot**
* **rugplot**

---

### üîπ distplot()

* Older Seaborn function
* Used to visualize data distribution
* Combines:

  * Histogram
  * KDE (Kernel Density Estimation)
  * Rug plot
* ‚ùå Deprecated (not used now)

---

### üîπ histplot()

* Modern replacement for distplot
* Mainly used for **histograms**
* KDE can be added optionally
* ‚úÖ Active and recommended
* ‚úÖ Introduced in Seaborn **0.11+**
**histplot = new & recommended**
---

## üîπ Joint Plot (`jointplot`)

* Used to visualize the **relationship between two quantitative variables**
* Shows:

  * **Scatter / hex / KDE plot** (center)
  * **Distribution of each variable** on the sides



---

## üîπ Rug Plot (`rugplot`)

* Displays **individual data points** along an axis
* Looks like small **tick marks**
* Helps identify **data concentration**


```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
sns.displot(df['total_bill'], bins = 10 , kde = True)
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], bins = 20, kde = True)
plt.subplot(1,2,2)
sns.histplot(df['tip'], bins = 20, kde = True)
```

```python
sns.jointplot(x = 'total_bill' , y = 'tip' , data = df, kind='scatter')
```

```python
sns.jointplot(x = 'total_bill' , y = 'tip' , data = df, kind='hex' )
```

```python
sns.jointplot(x = 'total_bill' , y = 'tip' , data = df, kind='reg' )
```

```python
sns.pairplot(df , hue='size', palette='rainbow') #hue means dividing the data into categories
```

```python
sns.rugplot(df['total_bill'])
```
